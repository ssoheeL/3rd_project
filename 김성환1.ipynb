```python
df = pd.read_csv("tips_data.csv")
df
```

```python
df['date'] = pd.to_datetime(df['date'], format='%Y%m%d')
print(df['date'])
```

```python
df.info()
```

```python
df['week'] = df['date'].dt.isocalendar().week
df

```

```python
wau = df.groupby(df['date'].dt.isocalendar().week)['fullVisitorId'].nunique()
print(wau)
```

```python
df['wau'] = df['date'].dt.isocalendar().week.map(wau)
df
```

```python
df['year_week'] = df['date'].dt.strftime('%Y-%U')
df.sort_values(by='year_week', inplace=True)
df
```

```python
wau = df.groupby(df['date'].dt.strftime('%Y-%U'))['fullVisitorId'].nunique()
df['wau'] = df['date'].dt.strftime('%Y-%U').map(wau)
df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
wau.plot(kind='line', marker='o', color='b')
plt.title('Weekly Active Users (WAU)')
plt.xlabel('Week')
plt.ylabel('WAU')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
